# Preprocessing of data

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
x = scaler.fit_transform(x)
x = pd.DataFrame(x)
```

```python
data = pd.concat([x, y, subject], axis=1) 
data.head()
```

# What is the final problem we have to solve ?

It is **Anamoly Detection combined with multilabel classification**

**Anamoly Detection** part will tell from the input features that whether or not the person parameters are normal or not.

**Multi-label classification** part will tell that what the person is doing and which person is doing

# Unsupervised Anamoly Detection

So since we have **No free lunch theorem** in Machine Learning, we have to try different methods for our data and see what fits best for it.

I will use the following two unsupervised anamoly detectors and use **Precision** as my evaluation metric.

* **Isolation Forest**
* **Autoencoders**

# Isolation Forest

```python
import pandas as pd
from sklearn.ensemble import IsolationForest
anomaly_model = IsolationForest(contamination=0.01)
anomaly_model.fit(x_train)
```

```python
c = 0
t = anomaly_model.predict(x_train)
for i in t:
    if i == -1:
        c = c + 1
print("Wrong predictions made on training data of size ",len(x_train)," = ",c)   
```

```python
c = 0
t = anomaly_model.predict(x_test)
for i in t:
    if i == -1:
        c = c + 1
print("Wrong predictions made on testing data of size ",len(x_test)," = ",c)   
```

# Autoencoders

```python
import pandas as pd
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Input, Dense
from tensorflow.keras.optimizers import Adam

input_dim = x_train.shape[1]
encoding_dim = 100

input_layer = Input(shape=(input_dim,))
encoder = Dense(encoding_dim, activation="relu")(input_layer)
decoder = Dense(input_dim, activation='sigmoid')(encoder)

autoencoder = Model(inputs=input_layer, outputs=decoder)

autoencoder.compile(optimizer=Adam(), loss='mean_squared_error')

autoencoder.fit(x_train, x_train, epochs=50, batch_size=32, validation_split=0.2)
```

```python
x_train_pred = autoencoder.predict(x_train)
mse = np.mean(np.power(x_train - x_train_pred, 2), axis=1)
threshold = np.quantile(mse, 0.99)
```

```python
def anomaly_detector(new_data):
    new_data_pred = autoencoder.predict(new_data)
    mse_new = np.mean(np.power(new_data - new_data_pred, 2), axis=1)
    return mse_new > threshold
```

```python
c = 0
t = anomaly_detector(x_train)
for i in t:
    if i == True:
        c = c + 1
print("Wrong predictions made on training data of size ",len(x_train)," = ",c)   
```

```python
c = 0
t = anomaly_detector(x_test)
for i in t:
    if i == True:
        c = c + 1
print("Wrong predictions made on testing data of size ",len(x_test)," = ",c)   
```

**So both the methods are performing almost equal in performance**

Since Autoencoder used distance metric which can be for high-dimensional data therefore I will prefer **Isolation Forest** Method

# What methods will not work and why ?

* **Regression:** It will not work as we have the predict something which is categorical in nature but Regression is a kind of technique which works well only for predicting data which is continuous in nature.   **Rejected**

* **KNN Classification:** Though this is a classification model it will not work in our case as our data has very high number of features (651 features) therefore the this model will suffer from **curse of dimensionality** and will not perform well on our data.   **Rejected**


# Multi-labelled Multi-class Classification

We will make a deep neural netwrok for this and try to different architechture for this

```python
from keras.models import Model
from keras.layers import Input, Dense
from keras.optimizers import Adam

input_shape = (x.shape[1],)

input_layer = Input(shape=input_shape)
q = Dense(512, activation='relu')(input_layer)
q = Dense(256, activation='relu')(q)
q = Dense(128, activation='relu')(q)
q = Dense(64, activation='relu')(q)

output_layer1 = Dense(6, activation='softmax', name='activity')(q)

model1 = Model(inputs=input_layer, outputs=output_layer1)

model1.compile(optimizer=Adam(), loss={'activity': 'categorical_crossentropy'},metrics=['accuracy'])

model1.summary()
```

```python
# Train the model
history = model1.fit(x_train, activity_train,epochs=50, batch_size=32, validation_split=0.2)
```

# Final Model after combining the above model

```python
def whether_employee_or_not(test_data):
    t = anomaly_model.predict(x_test)
    return t

def which_employee(test_data):
    predictions = model2.predict(x_test)
    return predictions

def which_activity(test_data):
    predictions = model1.predict(x_test)
    return predictions

def main_fn(test_data):
    activity_labels = ["WALKING","WALKING_UPSTAIRS","WALKING_DOWNSTAIRS","SITTING","STANDING","LAYING"]
    t = whether_employee_or_not(test_data)
    p1 = which_employee(test_data)
    p2 = which_activity(test_data)
    predicted_activities = []
    predicted_employees = []
    for i in p2:
        predicted_activities.append(np.argmax(i))
    for j in p1:
        predicted_employees.append(np.argmax(j))
    for i in range(len(t)):
        if(t[i] == -1):
            print("This actions does not seem to be of your company employee")
        else:
            print("Employee with employee id",predicted_employees[i]," is doing", activity_labels[predicted_activities[i]],"activity")
```
